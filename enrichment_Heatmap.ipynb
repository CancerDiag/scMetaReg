```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set_theme()
```

```python
eec_module6 = pd.read_csv("./EEC_results/module6_enrichment.tsv", sep = "\t")
eec_module6 = eec_module6.drop('Unnamed: 0', axis=1)
print(eec_module6.shape)

eec_module8 = pd.read_csv("./EEC_results/module8_enrichment.tsv", sep = "\t")
eec_module8 = eec_module8.drop('Unnamed: 0', axis=1)
print(eec_module8.shape)
```

```python
ov_module3 = pd.read_csv("./OV_results/module3_enrichment.tsv", sep = "\t")
ov_module3 = ov_module3.drop('Unnamed: 0', axis=1)
print(ov_module3.shape)

ov_module5 = pd.read_csv("./OV_results/module5_enrichment.tsv", sep = "\t")
ov_module5 = ov_module5.drop('Unnamed: 0', axis=1)
print(ov_module5.shape)
```

```python
cesc_module1 = pd.read_csv("./CESC_results/module1_enrichment.tsv", sep = "\t")
cesc_module1 = cesc_module1.drop('Unnamed: 0', axis=1)
print(cesc_module1.shape)

cesc_module4 = pd.read_csv("./CESC_results/module4_enrichment.tsv", sep = "\t")
cesc_module4 = cesc_module4.drop('Unnamed: 0', axis=1)
print(cesc_module4.shape)

cesc_module8 = pd.read_csv("./CESC_results/module8_enrichment.tsv", sep = "\t")
cesc_module8 = cesc_module8.drop('Unnamed: 0', axis=1)
print(cesc_module8.shape)
```

```python
tnbc_module2 = pd.read_csv("./BRCA_TNBC_results/module2_enrichment.tsv", sep = "\t")
tnbc_module2 = tnbc_module2.drop('Unnamed: 0', axis=1)
print(tnbc_module2.shape)

tnbc_module10 = pd.read_csv("./BRCA_TNBC_results/module10_enrichment.tsv", sep = "\t")
tnbc_module10 = tnbc_module10.drop('Unnamed: 0', axis=1)
print(tnbc_module10.shape)
```

```python
er_module2 = pd.read_csv("./BRCA_ER_results/module2_enrichment.tsv", sep = "\t")
er_module2 = er_module2.drop('Unnamed: 0', axis=1)
print(er_module2.shape)

er_module5 = pd.read_csv("./BRCA_ER_results/module5_enrichment.tsv", sep = "\t")
er_module5 = er_module5.drop('Unnamed: 0', axis=1)
print(er_module5.shape)

er_module6 = pd.read_csv("./BRCA_ER_results/module6_enrichment.tsv", sep = "\t")
er_module6 = er_module6.drop('Unnamed: 0', axis=1)
print(er_module6.shape)
```

```python
pathways = set(eec_module6["pathway_lst"].values).union(set(eec_module8["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(ov_module3["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(ov_module5["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(cesc_module1["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(cesc_module4["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(cesc_module8["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(tnbc_module2["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(tnbc_module10["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(er_module2["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(er_module5["pathway_lst"].values))
print(len(pathways))
pathways = pathways.union(set(er_module6["pathway_lst"].values))
print(len(pathways))
```

```python
heatmap_df = pd.DataFrame(pathways)
heatmap_df.columns = ["Metabolic_Pathways"]
heatmap_df["EEC_M6"] = ""
heatmap_df["EEC_M8"] = ""
heatmap_df["HGSOC_M3"] = ""
heatmap_df["HGSOC_M5"] = ""
heatmap_df["CC_M1"] = ""
heatmap_df["CC_M4"] = ""
heatmap_df["CC_M8"] = ""
heatmap_df["TNBC_M2"] = ""
heatmap_df["TNBC_M10"] = ""
heatmap_df["ER_M2"] = ""
heatmap_df["ER_M5"] = ""
heatmap_df["ER_M6"] = ""
for idx in heatmap_df.index:
    mp = heatmap_df.loc[idx ,"Metabolic_Pathways" ]
    if mp in eec_module6["pathway_lst"].values:
        heatmap_df.loc[idx,"EEC_M6"]=-np.log(eec_module6[eec_module6["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in eec_module8["pathway_lst"].values:
        heatmap_df.loc[idx,"EEC_M8"]=-np.log(eec_module8[eec_module8["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in ov_module3["pathway_lst"].values:
        heatmap_df.loc[idx,"HGSOC_M3"]=-np.log(ov_module3[ov_module3["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in ov_module5["pathway_lst"].values:
        heatmap_df.loc[idx,"HGSOC_M5"]=-np.log(ov_module5[ov_module5["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in cesc_module1["pathway_lst"].values:
        heatmap_df.loc[idx,"CC_M1"]=-np.log(cesc_module1[cesc_module1["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in cesc_module4["pathway_lst"].values:
        heatmap_df.loc[idx,"CC_M4"]=-np.log(cesc_module4[cesc_module4["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in cesc_module8["pathway_lst"].values:
        heatmap_df.loc[idx,"CC_M8"]=-np.log(cesc_module8[cesc_module8["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in tnbc_module2["pathway_lst"].values:
        heatmap_df.loc[idx,"TNBC_M2"]=-np.log(tnbc_module2[tnbc_module2["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in tnbc_module10["pathway_lst"].values:
        heatmap_df.loc[idx,"TNBC_M10"]=-np.log(tnbc_module10[tnbc_module10["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in er_module2["pathway_lst"].values:
        heatmap_df.loc[idx,"ER_M2"]=-np.log(er_module2[er_module2["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in er_module5["pathway_lst"].values:
        heatmap_df.loc[idx,"ER_M5"]=-np.log(er_module5[er_module5["pathway_lst"]==mp]["adjp_values"].values[0])
    if mp in er_module6["pathway_lst"].values:
        heatmap_df.loc[idx,"ER_M6"]=-np.log(er_module6[er_module6["pathway_lst"]==mp]["adjp_values"].values[0])

print(heatmap_df[(heatmap_df.iloc[:,1:] == "" ).all(1)].shape)
print(heatmap_df.shape)
heatmap_df.head()
```

```python
hm_df = heatmap_df.copy()
hm_df.replace(to_replace ="",value = 0, inplace= True)
hm_df.set_index("Metabolic_Pathways", inplace = True)
hm_df[hm_df > 12.5] = 12.5
```

```python
a = ['Carnitine shuttle (endoplasmic reticular)',
 'Carnitine shuttle (mitochondrial)',
 'Carnitine shuttle (peroxisomal)','C5-branched dibasic acid metabolism',
 'Ether lipid metabolism','Beta-alanine metabolism', 'Acyl-CoA hydrolysis',
 'Acylglycerides metabolism', 'Vitamin B6 metabolism','Pantothenate and CoA metabolism',
 'Terpenoid backbone biosynthesis', 'Fatty acid elongation','Ubiquinone synthesis',
 'Starch and sucrose metabolism',
 'Sulfur metabolism','Retinol metabolism', 
]
print(len(a))

a = ['Carnitine shuttle (endoplasmic reticular)',
 'Carnitine shuttle (mitochondrial)',
 'Carnitine shuttle (peroxisomal)']

print(len(a))
```

```python
filter_pathways = [ p for p in pathways if p not in a]
print(len(filter_pathways))
```

```python
# Main Plot
g = sns.clustermap(hm_df[hm_df.index.isin(filter_pathways)], cmap= "Oranges", 
                   linewidths=.5, 
                   yticklabels=True, figsize=(10,15),
                   cbar_pos=(-0.1, 0.5, 0.025, 0.1), 
                   dendrogram_ratio=0.0001, linecolor='black',
                   row_cluster=True,
                   cbar_kws={'label': '-log10(adj p-value)'},
                   col_cluster=True)
```
